# Supervised learning Model


```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as snp
%matplotlib inline
import pandas as pd
#importing necessary liabraries
```

```python
url="http://bit.ly/w-data"
#datset importing from url
```

```python
df=pd.read_csv(url)
#reading dataset
```

```python
df.head()
```

```python
#checking null values
df['Hours'].isnull().value_counts()
```

```python
df.describe()
#function for describing dataset
```

## plotting the data

```python
#plotting the columns of dataset
snp.scatterplot(x=df['Hours'],y=df['Scores'],data=df)
```

# checking co-relation

```python
df.corr(method='pearson')
#checking relation means strong co relation
```

```python
hours=df['Hours']
scores=df['Scores']

snp.distplot(hours)
snp.set_style('whitegrid')
```

```python
snp.distplot(scores)
```

```python

X=df.iloc[:, :-1].values
y=df.iloc[:,1]
```

# splitting data for training

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
#80% and 20% test training size
```

# Linear regression

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_predict=lr.predict(X_test)
```

```python
#line y=mx+c equa
m=lr.coef_
c=lr.intercept_
line=m*X+c
plt.scatter(X,y)
plt.plot(X,line)
plt.show()
```

```python
print(X_test)
```

```python
###calculating score
lr.score(X_test,y_test)
```

```python
#checking actual vs predicted 
df1=pd.DataFrame({'Actual':y_test,'predicted':y_predict})
df1
```

```python
a=np.array(y_test-y_predict)
print(a)
```

```python
snp.set_style('whitegrid')
snp.distplot(a)
plt.show()
```

```python
#testig with our own data
hours=9.25
s=lr.predict([[hours]])
print('if a student studies for {} hours then his/her score would be score {}% in exam: '.format(hours,s))
#print("the score is: ",out,"if you studied ",f, 'hours')
```

```python
# Model evaluation
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
print('the mean squared error is: ',mean_squared_error(y_test,y_predict))
print('the r2_score is: ',r2_score(y_test,y_predict))
```
